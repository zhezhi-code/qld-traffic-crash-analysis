## 2. Encode and Prepare Selected Features

This section performs label encoding, standardisation, train/test splitting, and SMOTE resampling for selected crash and road roughness features.

This cell is kept as a preparation step because it creates `X_train`, `X_test`, `y_train`, `y_test`, `X_train_resampled`, and `y_train_resampled` for later modelling.


```
data = df_cd

# Label Encoding and data stanlization
label_cols = ['Count_Casualty_Hospitalised', 'Count_Casualty_MedicallyTreated','NAASRARoughness', 'IRIRoughness']
le = LabelEncoder()

#Label Encoding
for col in label_cols:
    data[col] = le.fit_transform(data[col])

# extract feature（X）and target variable（y）
X = data[['Crash_Year', 'NAASRARoughness',
       'IRIRoughness']]
y = data['Crash_Severity']

# data standlization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# split dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
```
